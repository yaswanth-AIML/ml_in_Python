```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5],dtype=float)
y=np.array([3,5,7,9,11])
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
m=x.shape[0]
```

```python
print(x)
print(y)
```

```python
w=0.0
b=0.0
```

```python
def predict(x,w,b):
  return x*w+b
```

```python
print(predict(x,w,b))
```

```python
def compute_cost(x,y,w,b):
  y_pred=predict(x,w,b)
  cost=(1/m)*np.sum((y_pred-y)**2)
  return cost
```

```python
print(compute_cost(x,y,w,b))
```

compute gradients

```python
def compute_gradients(x,y,w,b):
  y_pred=predict(x,w,b)
  dw=(2/m)*np.sum((y_pred-y)*x)
  db=(2/m)*np.sum(y_pred-y)
  return dw,db
```

```python
print(compute_gradients(x,y,w,b))
```

```python
learning_rate=0.01
epochs=1000
for i in range(epochs):
  dw,db=compute_gradients(x,y,w,b)
  w=w-learning_rate*dw
  b=b-learning_rate*db
  if i%100==0:
    cost=compute_cost(x,y,w,b)
    print(f"epoch{i}:cost={cost:.4f}")
```

```python
print("learning rate:",w)
print("learning bias:",b)
```
